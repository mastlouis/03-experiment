```python
import numpy as np
import random
import pandas as pd

random.seed(9)

ndatapoints = 10
ndatasets = 20

d = pd.DataFrame(np.zeros((ndatasets, ndatapoints)))

i = 0
while i < ndatasets:
    # print('new one')
    for j in range(len(d.loc[0])):
        d[j][i] = random.random()

    # normalize
    s = sum(d.loc[i])
    for j in range(len(d.loc[i])):
      d[j][i] = d[j][i] / s * 100
    
    valid = True
    for j in range(ndatapoints):
        if d[j][i] < 3 or d[j][i] > 39:
            valid = False
            # print('min or max')
            break
        for k in range(ndatapoints):
            if k != j and abs(d[j][i] - d[k][i]) < .1:
                valid = False
                # print('less than 0.1 diff')
                break

    if valid:
        print(d.iloc[i], 'sum', sum(d.iloc[i]))
        i += 1
    # else:
    #   print('rejected', i)
    #   print(d[i])
```

```python
import numpy as np
import random
import pandas as pd

random.seed(9)

ndatapoints = 10
ndatasets = 20
num_neighbor_pairings = ndatasets / 2
d = pd.DataFrame(columns = 'row_index', 'comp_idx_1', 'comp_idx_2'])

for row in range(ndatasets):
    random_index = random_index = random.randint(0, ndatapoints)
    other_index = -1

    if (num_neighbor_pairings > 0 and random.randint(0,2) == 1) {
        # neighbors pairing
        other_index = (random_index + 1) % ndatapoints
        num_neighbor_pairings =- 1
    }
    else {
        # distanced pairing
        other_index = (random_index + random.randint(2,ndatapoints-2)) % ndatapoints
    }
    
    new_row = {'row_index':row, 'comp_idx_1':random_index, 'comp_idx_2':other_index}
    d.append(new_row)
# for every row in the dataset so 0 to 19,
# pick two indicies to compare :)
# 10 neighbors, 10 far 
```
